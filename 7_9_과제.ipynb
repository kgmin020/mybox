```python
# 1. 한글 폰트 설치 및 설정
!apt-get -qq -y install fonts-nanum

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager as fm

# 폰트 경로 설정
font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
fontprop = fm.FontProperties(fname=font_path)
plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False

# 2. 데이터 불러오기
df = pd.read_csv('/content/drive/MyDrive/dataset/Carseats.csv')

# 3. 수치형 변수만 선택
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

# 4. 상관계수 계산 및 히트맵 시각화
corr = df[numeric_cols].corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm',
            cbar=True, square=True,
            annot_kws={"fontsize":10, "fontproperties": fontprop})

plt.title('주요 수치형 변수 간 상관관계 히트맵', fontproperties=fontprop, fontsize=14)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 한글 폰트 설정 (이미 설치되어 있다 가정)
font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
fm.fontManager.addfont(font_path)
plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False

# 데이터 불러오기
df = pd.read_csv('/content/drive/MyDrive/dataset/Carseats.csv')

# 변수 조합 리스트: (독립변수, 종속변수)
var_pairs = [('Price', 'Sales'), ('Advertising', 'Sales'), ('Age', 'Sales'), ('Sales', 'Advertising')]

plt.figure(figsize=(16, 12))
for i, (x_var, y_var) in enumerate(var_pairs):
    plt.subplot(2, 2, i+1)
    sns.regplot(x=x_var, y=y_var, data=df,
                scatter_kws={'alpha':0.6, 'color':'skyblue'},
                line_kws={'color':'red'})
    plt.title(f'{x_var} → {y_var} 산점도 및 회귀선')
    plt.xlabel(x_var)
    plt.ylabel(y_var)
    plt.grid(True)

plt.tight_layout()
plt.show()
```
