**메타 데이터 기반 추천 알고리즘**

```python
import pandas as pd
import numpy as np

# 데이터 불러오기
movies = pd.read_csv('https://raw.githubusercontent.com/sd12832/MoViZ/master/src/tmdb_5000_movies.csv')
credits = pd.read_csv('https://raw.githubusercontent.com/sd12832/MoViZ/master/src/tmdb_5000_credits.csv')

# 컬럼 정리
credits.columns = ['id', 'title', 'cast', 'crew']
movies['id'] = movies['id'].astype(int)
df = movies.merge(credits, on='id')
```

```python
import ast

# 문자열을 실제 리스트(dict)로 변환
def parse_features(x):
    try:
        return [i['name'] for i in ast.literal_eval(x)]
    except:
        return []

df['genres'] = df['genres'].apply(parse_features)
df['keywords'] = df['keywords'].apply(parse_features)

# 상위 3명 배우만 추출
def get_cast(x):
    try:
        return [i['name'] for i in ast.literal_eval(x)[:3]]
    except:
        return []

df['cast'] = df['cast'].apply(get_cast)

# 감독만 추출
def get_director(x):
    try:
        for i in ast.literal_eval(x):
            if i['job'] == 'Director':
                return [i['name']]
        return []
    except:
        return []

df['crew'] = df['crew'].apply(get_director)
```

```python
# 모두 합쳐서 하나의 'soup' 컬럼 만들기
def clean_join(x):
    return ' '.join(x['genres'] + x['keywords'] + x['cast'] + x['crew'])

df['soup'] = df.apply(clean_join, axis=1)
```

```python
df['soup']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Count 기반 벡터화
count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(df['soup'])

# 코사인 유사도
cosine_sim = cosine_similarity(count_matrix, count_matrix)
```

```python
# 추천 함수
def get_recommendations(title, cosine_sim=cosine_sim):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    movie_indices = [i[0] for i in sim_scores]
    return df[['title_x', 'genres', 'cast', 'keywords']].iloc[movie_indices]
```

```python
get_recommendations('Avatar')
```

**모든 영화에 대해 평점을 매겨 최고 등급의 영화를 추천해주는 알고리즘 구현**

```python
import pandas as pd

# 데이터 불러오기
movies = pd.read_csv('https://raw.githubusercontent.com/sd12832/MoViZ/master/src/tmdb_5000_movies.csv')
credits = pd.read_csv('https://raw.githubusercontent.com/sd12832/MoViZ/master/src/tmdb_5000_credits.csv')

# 컬럼 정리 및 병합
credits.columns = ['id', 'title', 'cast', 'crew']
movies['id'] = movies['id'].astype(int)
df = movies.merge(credits, on='id')
```

```python
#평점 관련 통계값 정의
C = df['vote_average'].mean()            # 전체 평균 평점
m = df['vote_count'].quantile(0.9)       # 상위 10% 기준 최소 투표 수
```

```python
#IMDB 가중 평점 계산 함수 정의
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    return (v / (v + m)) * R + (m / (v + m)) * C
```

```python
#영어권/비영어권으로 나눠서 가중 평점 계산
# 영어권
en_movies = df[df['original_language'] == 'en'].copy()
en_movies = en_movies[en_movies['vote_count'] >= m]
en_movies['score'] = en_movies.apply(weighted_rating, axis=1)

# 비영어권
non_en_movies = df[df['original_language'] != 'en'].copy()
non_en_movies = non_en_movies[non_en_movies['vote_count'] >= m]
non_en_movies['score'] = non_en_movies.apply(weighted_rating, axis=1)
```

```python
# 영어권 TOP 10
print("영어권 영화 추천 Top 10")
print(en_movies[['title_x', 'vote_average', 'vote_count', 'score']].sort_values('score', ascending=False).head(10))

# 비영어권 TOP 10
print("\n비영어권 영화 추천 Top 10")
print(non_en_movies[['title_x', 'original_language', 'vote_average', 'vote_count', 'score']]
      .sort_values('score', ascending=False)
      .head(10))
```
